```python
import numpy as np

# 创建一维数组
arr_1d = np.array([1, 2, 3])

# 创建二维数组
arr_2d = np.array([[1, 2, 3], [4, 5, 6]])

# 创建三维数组
arr_3d = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])

print(arr_1d)
print(arr_2d)
print(arr_3d)

# 二维列表示例
list_2d = [
    [1, 2, 3],
    [4, 5, 6]
]

# 打印二维列表
print("二维列表:")
for row in list_2d:
    print(row)
```
