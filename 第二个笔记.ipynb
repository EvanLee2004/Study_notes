## ndarray数组创建

```python
list = [1,2,3]
arr = np.array(list,dtype = np.float64)
print(arr)
print(arr.ndim)#维度
```

```python
#预定义形状
#全0 全1 未初始化 固定值
#全0
arr = np.zeros((2,3),dtype = np.int64)
print(arr)
print(arr.dtype)
```

```python
arr = np.ones((5,8),dtype = np.int64)
print(arr)
```

```python
#未初始化
arr = np.empty((5,8),dtype = np.int64)
print(arr). 
```

```python
#其他预定义填充
arr = np.full((9,9),2021,dtype = np.int64)
print(arr)
```

```python
arr1 = np.zeros_like(arr)
print(arr1)
arr1 = np.ones_like(arr)
print(arr1)
arr1 = np.empty_like(arr)
print(arr1)
arr1 = np.full_like(arr,2026)
print(arr1)
```

```python
#等差数列
arr = np.arange(2,11,2)#不包含最后一个 11
print(arr)
```

```python
#等间隔数列
arr = np.linspace(1,10  ,5,dtype = int)#把前后长度均分成3个等间隔的数
print(arr)
 
```

```python
#对数间隔
arr = np.logspace(1,10,10,base = 2)#base:以谁为底，（开始，结束，个数）
print(arr)
 
```

## NumPy 函数与数学公式对应

| 函数          | 数学类型     | 公式/描述                  | 示例输出                  |
|---------------|--------------|----------------------------|---------------------------|
| `np.arange(start, stop, step)` | 等差数列    | a_n = a_1 + (n-1)*d       | [2, 4, 6, 8, 10]（d=2） |
| `np.linspace(start, stop, num)` | 等差数列    | a_n = a_1 + (n-1)*d       | [1, 3.25, 5.5, 7.75, 10]（d自动算） |
| `np.logspace(start, stop, num, base=b)` | 等比数列    | a_n = a_1 * r^(n-1)       | [1, 2, 4, 8]（r=2）     |

- **等差数列**：公差 d 固定。
- **等比数列**：公比 r 固定。

```python
#特殊矩阵的构造
#单位矩阵:主对角线上的元素为1，其余元素为0的方阵
arr =  np.eye(3,4,dtype = int)
print(arr)
```

```python
#对角矩阵：主对角线上是非0的数字，其他数字为0
arr = np.diag([5,2,3,4])
print(arr)
```

### 4 ndarray的数据类型
布尔类型 bool（对错）
整数类型 int uint
浮点数 float
复数 complex

```python
arr = np.array([1,0,2,0],dtype = np.bool)  #非0为True，0为False
print(arr)
```

```python
#索引(标注数组元素的位置的数字）与切片 
arr = np.random.randint(1,100,30)
print(arr)
print(arr[0])#第一个元素
print(arr[:])#所有元素
print(arr[2:5])#第三到第五个元素，左包右不包 n
print(arr[slice(2,5,3)])#第三到第五个元素，左包右不包,隔3切。一个
print(arr[arr>10])#布尔索引，通过布尔条件筛选，对应位置为True的元素会被选出
print(arr[(arr>10) & (arr<50)])#多条件筛选
```

```python
arr = np.random.randint(1,100,(4,8))
print(arr)
print(arr[(1,3)])#索引
print(arr[1,:])  
print(arr[2] [arr[2]>50])
print(arr[:,2])
```

```python
import numpy as np

#ndarray数组的运算
arr = np.array([1,2,3])
arr1 = np.array([4,5,6])
print(arr + arr1)#加减乘除一一对应位置运算

#ps:列表
list1 = [1,2,3]
list2 = [4,5,6]
print(list1 + list2)#列表相加是拼接，不是对应位置相加
```
